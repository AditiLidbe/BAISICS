5.Create an iterator class in python which that generates the fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n  # Number of terms
        self.a, self.b = 0, 1  # First two Fibonacci numbers
        self.count = 0  # Counter to track iterations

    def __iter__(self):
        return self  # Return the iterator object itself

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration  # Stop when the limit is reached

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b

        self.a, self.b = self.b, self.a + self.b  # Generate next term
        self.count += 1
        return self.b
```

7.Implement a genrator function that read a file line by line and yeild each line as a string.


```python
def read_file_line_by_line(file_path):
    file = open(file_path, 'r')
    try:
        line = ""
        while True:
            char = file.read(1)
            if not char and not line:
                break
            if char == "\n" or not char:
                yield line
                line = ""
            else:
                line += char
    finally:
        file.close()
```

9.Write a python program that uses map() to convert celsius into fahrenheit.


```python
celsius_temps = list(map(float, input("Enter temperatures in Celsius (comma-separated): ").split(',')))
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))
print(f"Temperatures in Fahrenheit: {fahrenheit_temps}")
```

10.Write a program that uses filter() to remove all the vowels from the string.


```python
# Function to remove vowels from a string using filter()
def remove_vowels(string):
    vowels = "aeiouAEIOU"  # Define vowels to be removed
    return "".join(filter(lambda char: char not in vowels, string))  # Filter out vowels

# Get input from the user
input_string = input("Enter a string: ")  # Allow user to enter a string
filtered_string = remove_vowels(input_string)  # Process the input
print(f"String without vowels: {filtered_string}")  # Print the result
```

11.Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.


```python
orders = [
    (34587, "Learning Python, Mark Lutz", 4, 40.95),
    (98762, "Programming Python, Mark Lutz", 5, 56.80),
    (77226, "Head First Python, Paul Barry", 3, 32.95),
    (88112, "Einfuhrung in Python3, Bernd Klein", 3, 24.99)
]

# Lambda function to calculate total price with adjustment if below 100
order_totals = list(map(lambda x: (x[0], x[2] * x[3] + (10 if x[2] * x[3] < 100 else 0)), orders))

# Print the order totals
print(order_totals)
```
